```python
import json
import requests
import pandas as pd
```

```python
c_symbols = ['BTC', 'ETH', 'LTC']
f_symbols = ['EUR', 'GBP']

def api_call(c_symbol, f_symbol):
    
    for c_sym in c_symbol:
        if c_sym == 'BTC':
            slug = 'bitcoin'
        elif c_sym == 'ETH':
            slug = 'ethereum'
        else:
            slug = 'litecoin'
        
        for f_sym in f_symbol:
            api_url = f'https://api.alternative.me/v1/ticker/{slug}/?convert={f_sym}'
            response = requests.get(api_url).json()
            print(json.dumps(response, indent=2))
            # here's where I want to create a variable with the iterables.
            #ex: {c_sym}_{f_sym} = response[0]['price_usd']
            # return {c_sym}_{f_sym}
api_call(c_symbols, f_symbols)
```
